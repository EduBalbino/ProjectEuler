```python
for x in range(2, 10):
    for y in range(10 * x + 1, 10 * (x + 1)):
        list_of_number[y].append(list_of_number[10 * x][1])
        list_of_number[y].append(list_of_number[y - (10 * x)][1])
for x in range(2, 10):
    list_of_number[100 * x].append(list_of_number[x][1])
    list_of_number[100 * x].append(list_of_number[100][1])
for x in range(1, 10):
    for y in range(100 * x + 1, 100 * (x + 1)):
        list_of_number[y].append(list_of_number[x][1])
        list_of_number[y].append(list_of_number[100][1])
        list_of_number[y].append('AND')
        list_of_number[y].extend(list_of_number[y - (100 * x)][1:])
```
