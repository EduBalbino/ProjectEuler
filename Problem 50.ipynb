```python
def is_prime(n):
    if n < 2: 
         return False;
    if n % 2 == 0:             
         return n == 2  # return False
    k = 3
    while k*k <= n:
         if n % k == 0:
            return False
         k += 2
    return True
```

```python
def sublists(seq, n):
    return [a for i in range(0,len(seq)) if len(a:=seq[i:i+n]) == n]
```

```python
num = 1000001
iterable_primes = [x for x in range(num) if is_prime(x)]
search_primes = {x for x in range(num - 100000, num) if is_prime(x)}
```

```python
def p_50():
    for n in reversed(range(21, 550)):
        for x in sublists(iterable_primes, n):
            if (a:=sum(x)) in search_primes:
                return [a, len(x), x]
print(p_50())
```
