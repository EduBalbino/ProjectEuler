```python
def is_prime(n):
    if n < 2: 
         return False;
    if n % 2 == 0:             
         return n == 2  # return False
    k = 3
    while k*k <= n:
         if n % k == 0:
            return False
         k += 2
    return True
```

```python
listadenumeros = [1]
lista_de_primos = [False]
a = 0
b = 1

for y in range(2, 2*50000 + 1, 2):
    copiaint = int(listadenumeros[-1])
    listadenumeros.extend([copiaint + y, copiaint + 2*y, copiaint + 3*y])
    lista_de_primos.extend(list(map(is_prime, listadenumeros[-4:])))
    a += lista_de_primos[-4:].count(True)
    b += lista_de_primos[-4:].count(False)
    if a / b < 0.1:
        print((len(listadenumeros)-1)//4)
        print()
        print(listadenumeros[-50:])
        print()
        print(lista_de_primos.count(True)/lista_de_primos.count(False))
        break
```
